## Q1

```python
Education = df[(df['שם רמה 2'] == 'חינוך')]
Education
```

```python
def education_budget(year):
    specificYear = Education[(Education['שנה'] == year)]
    budgetDF = specificYear[(specificYear['הוצאה נטו']<0)]
    budget = budgetDF['הוצאה נטו'].sum()
    #specificYear = specificYear['הוצאה/הכנסה']
    return budget
```

```python
year = 1997
print("the eductaion budget in", year, "was", int(-1*education_budget(year)), "shekels")
year = 2005
print("the eductaion budget in", year, "was", int(-1*education_budget(year)), "shekels")
year = 2014
print("the eductaion budget in", year, "was", int(-1*education_budget(year)), "shekels")
year = 2021
print("the eductaion budget in", year, "was", int(-1*education_budget(year)), "shekels")
```

## Q2

```python
def security_budget_ratio(year):
    specificYear = df[(df['שנה'] == year)]
    security = specificYear[(specificYear['שם רמה 2'] == 'בטחון')]
    security = security[security['הוצאה נטו'] < 0]
    security_budget = float(security['הוצאה נטו'].sum())
    #print(security_budget)
    specificYear = specificYear[specificYear['הוצאה נטו'] < 0]
    total_budget = float(specificYear['הוצאה נטו'].sum())
    #print(total_budget)
    return (security_budget/total_budget)*100
```

```python
year = 2017
print("the seucrity budget in", year, "was", security_budget_ratio(year), "percentages of the total budget")
year = 2000
print("the seucrity budget in", year, "was", security_budget_ratio(year), "percentages of the total budget")
year = 2005
print("the seucrity budget in", year, "was", security_budget_ratio(year), "percentages of the total budget")
year = 2020
print("the seucrity budget in", year, "was", security_budget_ratio(year), "percentages of the total budget")
```

## Q3

```python
def largest_budget_year(office):
    officeTable = df[(df['שם רמה 2'] == office)]
    years = df["שנה"].unique()
    max = 0
    maxYear = 1997
    for year in years:
        yearTable = officeTable[(officeTable['שנה'] == year)]
        yearTable = yearTable[yearTable['הוצאה נטו'] < 0]
        budget = -1*(yearTable['הוצאה נטו'].sum())
        if budget > max:
            max = budget
            maxYear = year
    return maxYear
```

```python
offices = df["שם רמה 2"].unique()
print(offices)
```

```python
office = 'חינוך'
print("The max budget of the ", office, "office was in ", largest_budget_year(office))
office = 'בטחון'
print("The max budget of the ", office, "office was in ", largest_budget_year(office))
office = 'אנרגיה'
print("The max budget of the ", office, "office was in ", largest_budget_year(office))
office = 'חקלאות'
print("The max budget of the ", office, "office was in ", largest_budget_year(office))
```

## Mine Query

```python
#באילו משרדים הייתה בשנה מסוימת עלייה בתקציב לעומת השנה שלפני

def bigger_budget (year):
    officesToReturn = []
    if year==1997:
        print("don't have the data")
        return []
    this_year = df[(df['שנה'] == year)]
    prev_year = df[(df['שנה'] == year-1)]
    offices = df["שם רמה 2"].unique()
    #offices = ['חינוך', 'בטחון', 'אנרגיה']
    for office in offices:
        #print(office)
        officeThisYear = this_year[(this_year['שם רמה 2'] == office)]
        officeThisYear = officeThisYear[officeThisYear['הוצאה נטו'] < 0]
        this_year_budget = -1*(officeThisYear['הוצאה נטו'].sum())
        #print(this_year_budget)
        officePrevYear = prev_year[(prev_year['שם רמה 2'] == office)]
        officePrevYear = officePrevYear[officePrevYear['הוצאה נטו'] < 0]
        prev_year_budget = -1*(officePrevYear['הוצאה נטו'].sum())
        #print(prev_year_budget)
        if this_year_budget > prev_year_budget:
            officesToReturn.append(office)
    return officesToReturn
```

```python
year = 2017
print("The following offices had biggest budget in", year, "comapre to", year-1, ": ", bigger_budget(year))
year = 2000
print("The following offices had biggest budget in ", year, "comapre to ", year-1, ": ", bigger_budget(year))
year = 2021
print("The following offices had biggest budget in ", year, "comapre to ", year-1, ": ", bigger_budget(year))
year = 2010
print("The following offices had biggest budget in ", year, "comapre to ", year-1, ": ", bigger_budget(year))
```
