```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# This function is based on a similar solution I found in the internet 
def plotIntersection(range, f, g):
    plt.plot(range, f, '-')
    plt.plot(range, g, '-')

    intersectionsPoints = np.argwhere(np.diff(np.sign(f - g))).flatten()
    plt.plot(range[intersectionsPoints], f[intersectionsPoints], 'ro')
    plt.grid()
    plt.show()
```

```python
f = lambda x : x**2
g = lambda x : x+10
plotIntersection(np.linspace(-10, 10, 1000), f(x), g(x))
```

```python
f = lambda x : np.sin(x)
g = lambda x : 0.2*x
plotIntersection(np.linspace(-10, 10, 1000), f(x), g(x))
```

```python
f = lambda x : x
g = lambda x : -x
plotIntersection(np.linspace(-10, 10, 1000), f(x), g(x))
```

```python
f = lambda x : np.sin(x)
g = lambda x : np.cos(x)
plotIntersection(np.linspace(-10, 10, 1000), f(x), g(x))
```

```python
f = lambda x : (x**2)+3
g = lambda x : -(x**2)-3
plotIntersection(np.linspace(-10, 10, 1000), f(x), g(x))
```
